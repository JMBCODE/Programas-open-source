```python
#gráfico da função f(x) = 1/x
import matplotlib.pyplot as plt
 
x = [1/8]
y = [8]
auxx = x[-1] #pega o ultimo
auxy = y[-1] #pega o ultimo
while auxx <= 8:
  auxx = auxx * 2
  auxy = auxy / 2
  x.append(auxx)
  y.append(auxy)
  auxx = x[-1] #pega o ultimo
  auxy = y[-1] #pega o ultimo
 
plt.plot(x, y)
 
plt.show()
```

```python
#gráfico da função f(x) = x²
import matplotlib.pyplot as plt
 
x = [-15]
aux = x[-1] * x[-1]
y = [aux]
auxx = x[-1] #pega o ultimo
auxy = y[-1] #pega o ultimo
while auxx < 15:
  auxx = auxx + 1
  auxy = auxx # pra pegar 1, 2, 3... e não 1,4,9...
  auxy = auxy * auxy
  x.append(auxx)
  y.append(auxy)
  auxx = x[-1] #pega o ultimo
  auxy = y[-1] #pega o ultimo
 
#plt.figure(figsize=(7, 7))
plt.plot(x, y)
 
plt.show()
```

```python
 #gráfico da função f(x) = (x + 1) / (x - 1)
import matplotlib.pyplot as plt
import numpy as np
 
x = [2]
aux = (x[-1] + 1) / (x[-1] - 1)
y = [aux]
auxx = x[-1] #pega o ultimo
auxy = y[-1] #pega o ultimo
while auxx < 12:
  auxx = auxx + 0.001
  auxy = auxx # pra pegar 1, 2, 3... e não 1,4,9...
  auxy = (auxy + 1) / (auxy - 1)
  x.append(auxx)
  y.append(auxy)
  auxx = x[-1] #pega o ultimo
  auxy = y[-1] #pega o ultimo
 
#plt.figure(figsize=(5, 4))
plt.plot(x, y)
#plt.xticks(np.arange(min(x), max(x)+1, 1.0))
#plt.yticks(np.arange(min(y), max(y)+1, 1.0))
plt.show()
```
